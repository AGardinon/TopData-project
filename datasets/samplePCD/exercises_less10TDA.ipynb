```python
import numpy as np
from ripser import ripser
from persim import plot_diagrams
from persim import bottleneck_matching

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# compute diagram
# ok but how is it done by the cose?
n_set = 0
data = np.loadtxt("./datasets/data_0{}.txt".format(n_set), skiprows=1)
diagrams = ripser(data)['dgms']
plot_diagrams(diagrams, show=True, title="data_0{}".format(n_set))
```

```python
for i in range(5):
    n_set = i
    data = np.loadtxt("./datasets/data_0{}.txt".format(n_set), skiprows=1)
    diagrams = ripser(data, maxdim=1)['dgms']
    
    fig, ax = plt.subplots(1,2, figsize=(2 * 6, 1 * 5), dpi=100)
    ax.flatten()
    ax[0].scatter(data[:,0], data[:,1])
    ax[0].set_title("data_0{}".format(n_set))
    ax[0].set_xlabel('X')
    ax[0].set_ylabel('Y')
    plot_diagrams(diagrams, show=True, title="diagram_0{}".format(n_set), ax=ax[1])
    fig.tight_layout()
```
